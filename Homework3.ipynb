```python
#Q1-List all rows and columns of student table
txt='''
Select *
From Student
'''
Q1 =  psql.sqldf(txt)
Q1
```

```python
#Q2-List student first name and last name
txt='''
Select StdFirstName,StdLastName
From Student
'''
Q2 =  psql.sqldf(txt)
Q2
```

```python
#Q3-List student first name and last name whose major is 'IS'
txt='''
Select StdFirstName,StdLastName
From Student
Where StdMajor='IS'
'''
Q3 =  psql.sqldf(txt)
Q3
```

```python
#Q4-List student first name and last name whose major is 'ACCT' and student class is sophomere ('SO')
txt='''
Select StdFirstName,StdLastName
From Student
Where StdMajor='ACCT' and StdClass='SO'
'''
Q4 =  psql.sqldf(txt)
Q4
```

```python
#Q5-List all student with student GPA is below 3
txt='''
Select *
From Student
Where StdGpa<3
'''
Q5 =  psql.sqldf(txt)
Q5
```

```python
#Q6-List all student with student GPA is between 3 and 3.5 (3 and 3.5 is included)
txt='''
Select *
From Student
Where StdGpa>=3 and StdGpa<=3.5
'''
Q6 =  psql.sqldf(txt)
Q6
```

```python
#Q7-List unique student major
txt='''
Select Distinct(StdMajor)
From Student

'''
Q7 =  psql.sqldf(txt)
Q7
```

```python
#Q8-List faculty first name and last name whose rank is ASST or ASSC
txt='''
Select FacFirstName,FacLastName
From Faculty
Where FacRank=('ASST') or FacRank=('ASSC')
'''
Q8 =  psql.sqldf(txt)
Q8
```

```python
#Q9-List faculty first name and last name whose rank is not  ASST and ASSC
txt='''
Select FacFirstName, FacLastName
From Faculty
Where FacRank NOT IN ('ASST','ASSC')
'''
Q9 =  psql.sqldf(txt)
Q9
```

```python
#Q10-List the courses containing the string 'BUSINESS' in the course description.
txt='''
Select *
From Course
Where CrsDesc LIKE '%BUSINESS%'
'''
Q10 =  psql.sqldf(txt)
Q10
```
